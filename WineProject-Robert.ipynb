```python
#Import Dependancies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load in Data as CSV
wine_df = pd.read_csv("winemag-data-130k-v2.csv", encoding = "ISO-8859-1")
dropped_wine_df = wine_df.dropna(subset = ['price'])
print(len(dropped_wine_df))
```

```python
dropped_wine_df.describe()
```

```python
dropped_wine_df
```

```python
x_axis = dropped_wine_df['points']
price = dropped_wine_df['price']
```

```python
plt.xlim(80, 100)
plt.ylim(0, 500)

plt.title("Dollars to Points")
plt.xlabel("points")
plt.ylabel("$$$")

#Plotting all wines under $500
plt.scatter(x_axis, price, marker="o", color="red", s=.05)
```

```python
dropped_wine_df["UtilityMax"] = dropped_wine_df["points"]/dropped_wine_df["price"]
dropped_wine_df["UtilityMax"]
```

```python

BB_df = dropped_wine_df.sort_values("UtilityMax", ascending= False)
Top100_BB = BB_df.head(100)
NewDF = Top100_BB[["title", "designation","variety","points","price","UtilityMax","winery","country", "province"]]
NewDF
```
